# SerpAPI Web Search

```python
import os
import serpapi
from dotenv import load_dotenv
```

```python
load_dotenv()
```

```python
SERP_API_KEY = os.getenv("SERP_API_KEY")
```

```python
serp_queries = [
    'site:greenhouse.io ("AI engineer" OR "LLM engineer" OR "Machine Learning Engineer") (remote OR "remote-friendly" OR worldwide OR APAC OR Asia OR "South Korea") (agentic OR "AI agent" OR "agentic systems" OR RAG OR "retrieval-augmented" OR LoRA OR QLoRA OR "fine-tune" OR "Hugging Face" OR "open-source") (junior OR entry OR internship OR mid)',
    'site:lever.co intitle:careers ("agentic" OR "AI agent" OR "RAG" OR "LoRA" OR "QLoRA" OR "fine-tune" OR "Hugging Face") (remote OR worldwide OR Asia OR APAC OR "South Korea") (junior OR intern OR mid OR "early career")',
    'inurl:/careers ("AI engineer" OR "LLM engineer" OR "AI integration" OR "LLM engineering") (remote OR "remote-friendly" OR global OR Asia OR APAC OR "South Korea") ("Hugging Face" OR "open-source" OR transformers OR LoRA OR QLoRA OR RAG)',
    '(inurl:jobs OR inurl:careers OR intitle:jobs) ("agentic systems" OR "AI agents" OR agents) (remote OR worldwide OR Asia OR APAC OR "South Korea") (junior OR internship OR mid)',
    'site:workable.com ("AI engineer" OR "LLM" OR "LLM engineer") (remote OR APAC OR Asia OR "South Korea" OR worldwide) (LoRA OR QLoRA OR RAG OR "retrieval-augmented" OR "Hugging Face") (junior OR mid OR intern)',
]

exa_queries = [
    'Find remote junior or mid AI engineer roles on company careers or ATS pages (Greenhouse, Lever, Workable, Greenhouse boards) focused on agentic systems, AI agents, RAG pipelines, LoRA/QLoRA fine-tuning, Hugging Face or open-source LLMs — worldwide or APAC including South Korea',
    'Company careers pages hiring interns or early-career LLM engineers for RAG pipelines, LoRA/QLoRA fine-tuning, agentic AI, or AI-driven app integration; remote-friendly roles accepting applicants from Asia/APAC or global',
    'Search for remote-friendly junior AI engineer jobs on in-house careers pages or static ATS pages (Greenhouse, Lever, Workable) that mention AI agents, agentic systems, retrieval-augmented generation, LoRA, QLoRA, Hugging Face or open-source models',
    'List mid-level or internship openings at companies with direct careers pages for LLM engineering, agentic systems, RAG, fine-tuning (LoRA/QLoRA), model integration, remote worldwide or Asia/APAC-friendly',
    'Find company job posts (careers/, jobs/ or boards.greenhouse.io) for remote LLM engineer or AI agent engineer, focusing on open-source Hugging Face models, LoRA/QLoRA fine-tuning, RAG, accepting applicants in Asia including South Korea',
]
```

```python
import itertools
import time

client = serpapi.Client(api_key=SERP_API_KEY)
all_results = []
for i, q in enumerate(serp_queries):
    results = client.search({
        "engine": "google", 
        "google_domain": "google.com",
        "hl": "en", 
        "q": q
    })

    
    all_results.append(results)
    print(f"Search Interation {i + 1} Finished")
    time.sleep(1)

# Flattten all_results into flat_results. This should be returned in the real code 
# (falttened raw data and convert it into Job data schema in Normalize class later on)
flat_results = [
    {**item, "searched_via": "serp"} # Add a new key-vaue into the existing dict
    for item in itertools.chain.from_iterable(
        r.get("organic_results", []) for r in all_results
    )
]
```

```python
exa_client = Exa(api_key=EXA_API_KEY)
```
